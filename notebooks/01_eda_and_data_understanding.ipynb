---------------------------------------------------------------------------------------------------------------------------------------------

**Credit history vs Loan Status**

If someone has paid loans properly before, they’re trusted more.

So we expect:

    Good credit history → more approvals

    Bad or missing credit history → more rejections

If this isn’t true, that’s shocking.

Credit_History is a proxy for prior repayment behavior and is typically the strongest univariate predictor in credit risk datasets.

We examine conditional approval/default rates.

---------------------------------------------------------------------------------------------------------------------------------------------

```python
df.groupby('Credit_History')['Loan_Status'].value_counts(normalize = True)
```

---------------------------------------------------------------------------------------------------------------------------------------------

**What does this tell us?**

People with a missing credit history (0.0) have a very high rejection rate (0.921 ~ 92%) and a very low success rate (0.0786 ~ 7.8%).
People with a credit history have a mixed proportion with a moderate high success rate (0.795 ~ 79.5%) and a moderate rejection rate (0.204 ~ 20.4%).

**Loan Amount vs Loan Status**

Loan amount represents exposure:

    Bigger loans → higher potential loss

Banks are more cautious

---------------------------------------------------------------------------------------------------------------------------------------------

```python
df.groupby('LoanAmount')['Loan_Status'].value_counts(normalize = True)
```

```python
df.groupby('Loan_Status')['LoanAmount'].median()
```

---------------------------------------------------------------------------------------------------------------------------------------------

**What does this tell us?**

Rejected loans typically have higher loan amounts.

---------------------------------------------------------------------------------------------------------------------------------------------

**Income vs Loan Status**

Income seems important intuitively, but raw income is often noisy.

What we will be checking for?

Median applicant and coapplicant income by loan status

---------------------------------------------------------------------------------------------------------------------------------------------

```python
df.groupby('Loan_Status')[['ApplicantIncome', 'CoapplicantIncome']].median()
```

---------------------------------------------------------------------------------------------------------------------------------------------

**What does this tell us?**

Income differences are often small or inconsistent

Confirms income alone is not enough to judge risk
